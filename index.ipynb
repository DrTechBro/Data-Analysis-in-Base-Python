### Python Data Types and Control Structures

In this lesson, our data is stored in data structures that are built into the Python language: numbers, strings, lists, dictionaries, etc. For example, if we have this table of data:

| color  | number |
| ------ | ------ |
| green  | 7      |
| red    | 2      |
| orange | 1      |

A standard way we might represent that table would be a list of dictionaries, where each dictionary represents a row and has the keys `color` and `number`:

````python
info_table = [
    {"color": "green", "number": 7},
    {"color": "red", "number": 2},
    {"color": "orange", "number": 1}
]

type(info_table)
````

Then if we wanted to print out all of the values associated with the `color` keys, the logic would look something like this:

````python
for row in info_table:
    print(row["color"])
````

As more knowledge is gained, that same task (printing the list of colors) might look something like this, using the `pandas` library:

````python
import pandas as pd
df = pd.DataFrame(info_table)
type(df)

### import pandas as pd: 
### This line imports the pandas library and aliases it as pd, allowing you to refer to pandas functions and objects using the shorter alias pd.

### df = pd.DataFrame(info_table): 
### This line creates a DataFrame object named df by passing the info_table variable to the pd.DataFrame() constructor. A DataFrame is a two-dimensional labeled data structure with columns that can contain different data types (similar to a spreadsheet or SQL table). The info_table variable should be a data structure like a list of lists, a dictionary, or another DataFrame.


### type(df): This line uses the type() function to determine the type of the object stored in the variable df. It then prints the type of the object, indicating that df is a DataFrame.
````

````python
print(df["color"])
````

In the base Python example, we had a variable `info_table` which was type `list` (a built-in Python type), whereas in the pandas example, we had a variable `df` which was type `pandas.core.frame.DataFrame` (a custom type from the pandas library).

Then to print out the colors, in the base Python example we had a `for` loop and then code to extract the information from each individual dictionary. The string values got printed out without any additional markup. Then in the pandas example we did not have to use a loop (hint: this is *broadcasting* like we saw previously with NumPy) and also we had some extra markup where it printed out the index values as well as `Name: color, dtype: object`.

In this lesson you will practice performing data analysis with the first (base Python) format, before eventually learning how to use pandas instead.

### CSV Files

The first major file type we will explore is CSV (comma-separated value). For example, this code loads a dataset of 5000 apple trees in an orchard, where the column `yield` represents the total pounds of apples yielded by that tree for a given year.

You can think of the table as looking like this, except that it is thousands of lines long:

| yield              |
| -----------------  |
| 39.741234289561504 |
| 39.872054738763474 |
| 44.33116416558151  |
| 46.6006230827385   |
| 40.694984210927196 |
| 40.96981882686812  |

In relatively few lines of code, we can perform some key data analysis tasks: counting, measuring central tendency, finding the maximum, and finding the minimum.

````python
with open("data/apple_orchard.csv") as f:
    apple_orchard_data = list(csv.DictReader(f))
    apple_tree_yields = [float(x["yield"]) for x in apple_orchard_data]
    print("Total apple tree yields:", len(apple_tree_yields))
    print("Average apple tree yield:", sum(apple_tree_yields) / len(apple_tree_yields))
    print("Maximum apple tree yield:", max(apple_tree_yields))
    print("Minimum apple tree yield:", min(apple_tree_yields))
````

#### This code reads data from a CSV file named "apple_orchard.csv", processes it, and prints out some statistics related to apple tree yields. Here's a breakdown of what each part does:

1. ##### with open("data/apple_orchard.csv") as f::

This line opens the "apple_orchard.csv" file located in the "data" directory using a context manager (with statement). It opens the file in read mode ('r').

The file object is assigned to the variable f.

2. ##### apple_orchard_data = list(csv.DictReader(f)):

This line reads the contents of the CSV file using the csv.DictReader class. 

It reads each row of the CSV file as a dictionary where the keys are column headers and the values are the corresponding values in each row.

The result is converted into a list of dictionaries using the list() function and assigned to the variable apple_orchard_data.

3. ##### apple_tree_yields = [float(x["yield"]) for x in apple_orchard_data]:

This line extracts the "yield" values from each dictionary in apple_orchard_data and converts them to floating-point numbers.

It uses a list comprehension to iterate over each dictionary (x) in apple_orchard_data and extracts the "yield" value from each dictionary.

The resulting list of yields is assigned to the variable apple_tree_yields.

4. ##### print("Total apple tree yields:", len(apple_tree_yields)):

This line prints the total number of apple tree yields by calculating the length of the apple_tree_yields list using the len() function.

5. ##### print("Average apple tree yield:", sum(apple_tree_yields) / len(apple_tree_yields)):

This line calculates and prints the average apple tree yield by summing all the yields in apple_tree_yields using the sum() function and then dividing by the total number of yields.

6. ##### print("Maximum apple tree yield:", max(apple_tree_yields)):

This line prints the maximum apple tree yield by finding the maximum value in apple_tree_yields using the max() function.

7. ##### print("Minimum apple tree yield:", min(apple_tree_yields)):

This line prints the minimum apple tree yield by finding the minimum value in apple_tree_yields using the min() function.

#### Overall, this code reads apple tree yield data from a CSV file, calculates some statistics (total, average, maximum, and minimum yields), and prints them out.

### JSON Files

The second major file type we will explore is JSON (JavaScript object notation). For example, this code loads a dataset of interactions between Twitter users, where each user is represented as a "node" and when one tweets at another that connection is represented as a "link".

You can think of this data as looking like this, although this version is truncated:

```

  'links': [
    {'source': '347457291', 'target': '232762581'},
    {'source': '347457291', 'target': '119706266'},
    {'source': '347457291', 'target': '421509544'},
    {'source': '232762581', 'target': '222957350'},
    {'source': '232762581', 'target': '21648607'},

    {'source': '232762581', 'target': '59974294'},
```

In relatively few lines of code, we can count the number of users and find the users who were "sources" (the user initiating a tweet that tags someone else) most often and "targets" (the user being tagged in a tweet) most often.

````python
def build_freq_table(key, links):
    table = {}
    for link in links:
        user = link[key]
        table[user] = table.get(user, 0) + 1
    return table

def print_top_5(table):
    for k, v in sorted(table.items(), key=lambda x: x[1], reverse=True)[:5]:
        print(f"User {k}\t| {v} Tweets")

with open("data/twitter_graph.json") as f:
    twitter_data = json.load(f)
    print(f"There are {len(twitter_data['nodes'])} users in this dataset")
    print()
    
    links = twitter_data["links"]
    sources = build_freq_table("source", links)
    targets = build_freq_table("target", links)
        
    print("Top 5 Sources:")
    print_top_5(sources)
    print()
    print("Top 5 Targets:")
    print_top_5(targets)
````

This code reads data from a JSON file named "twitter_graph.json", processes it, and prints out some statistics related to Twitter users' interactions. Here's a breakdown of what each part does:

1. **`build_freq_table(key, links)` function:**
   - This function takes two arguments: `key`, which specifies the key in each link dictionary to use for frequency counting, and `links`, which is a list of dictionaries representing links between Twitter users.
   - Inside the function, it initializes an empty dictionary `table` to store the frequency counts.
   - It then iterates over each link in the `links` list. For each link, it extracts the value corresponding to the specified `key` and increments the count in the `table` dictionary for that value.
   - Finally, it returns the `table` dictionary containing the frequency counts.

2. **`print_top_5(table)` function:**
   - This function takes a dictionary `table` as input, where keys represent users and values represent frequency counts.
   - Inside the function, it sorts the dictionary items based on the frequency counts in descending order and selects the top 5 items.
   - It then iterates over the top 5 items and prints the user ID and the corresponding tweet count.

3. **Opening the JSON file:**
   - The code opens the "twitter_graph.json" file located in the "data" directory using a context manager (`with` statement) and assigns the file object to the variable `f`.

4. **Loading JSON data:**
   - It reads the contents of the JSON file using the `json.load()` function and parses it into a Python dictionary `twitter_data`.

5. **Extracting links data:**
   - It extracts the "links" data from the `twitter_data` dictionary and assigns it to the `links` variable.

6. **Calculating frequency of source users:**
   - It calls the `build_freq_table()` function with `"source"` as the key argument and `links` as the list of links. It calculates the frequency of each source user in the links and stores the results in the `sources` dictionary.

7. **Calculating frequency of target users:**
   - It calls the `build_freq_table()` function with `"target"` as the key argument and `links` as the list of links. It calculates the frequency of each target user in the links and stores the results in the `targets` dictionary.

8. **Printing statistics:**
   - The code prints the total number of users in the dataset, followed by the top 5 sources (users who initiated interactions) and top 5 targets (users who received interactions), using the `print_top_5()` function.
